```python
import pandas as pd
```

```python
superstore_df=pd.read_csv("Sample - Superstore.csv",encoding='latin-1')
```

```python
superstore_df=superstore_df.dropna()
```

```python
superstore_df.isnull().sum()
```

```python
superstore_df.describe()
```

```python
superstore_df.drop_duplicates(inplace=True)
```

```python
superstore_df['Ship Date']=pd.to_datetime(superstore_df['Ship Date'])
superstore_df['Order Date']=pd.to_datetime(superstore_df['Order Date'])
```

```python
superstore_df['Ship Date']
```

```python
region=superstore_df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
```

```python
category=superstore_df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

```python
ship_mode=superstore_df.groupby('Ship Mode')['Sales'].sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
region.plot(kind='bar')
plt.title("Sales by region")
plt.xlabel("region")
plt.ylabel("Sales")
plt.show()
```

```python
category.plot(kind='bar')
plt.title("Sales by category")
plt.xlabel("category")
plt.ylabel("Sales")
plt.show()
```

```python
ship_mode.plot(kind='bar')
plt.title("Sales by ship_mode")
plt.xlabel("ship_mode")
plt.ylabel("Sales")
plt.show()
```

```python
superstore_df['Month'] = superstore_df['Order Date'].dt.month
```

```python
sales_monthly = superstore_df.groupby('Month')['Sales'].sum()

sales_monthly.plot(kind='line' , figsize=(12,6), marker='o')
plt.title("Sales trend over time monthly")
plt.xlabel("month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
sales_monthly.max()
```

```python
region_profit=superstore_df.groupby('Region')['Profit'].sum().sort_values(ascending=False)
```

```python
category_profit=superstore_df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
```

```python
region_profit.plot(kind='bar')
plt.title("profit  by region")
plt.xlabel("region")
plt.ylabel("Profit")
plt.show()
```

```python
category_profit.plot(kind='bar', color='orange')
plt.title("profit  by category")
plt.xlabel("category")
plt.ylabel("Profit")
plt.show()
```

```python
loss_products = superstore_df[superstore_df['Profit'] < 0]
print(loss_products[['Product Name', 'Profit']].sort_values(by='Profit').head(15))
loss_products.groupby('Product Name')['Profit'].sum().sort_values().head(10).plot(kind='bar', color='darkred')
plt.title("Top 10 Products by Total Loss Amount")
plt.ylabel("Total Loss ($)")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(superstore_df['Sales'], superstore_df['Profit'], alpha=0.5)
plt.xscale('log') 
plt.title("Sales vs Profit")   
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.axhline(0, color='red', linestyle='--',linewidth=1)
plt.show()  
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=superstore_df, x='Sales', y='Profit', hue='Category', alpha=0.6)
plt.title("Sales vs Profit by Category")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.axhline(0, color='black', linewidth=1)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=superstore_df, x='Sales', y='Profit', hue='Category', size='Discount', sizes=(20, 200), alpha=0.6)
plt.title("Sales vs Profit (Size = Discount)")
plt.axhline(0, color='black', linewidth=1)
plt.show()
```

```python
top_sales = superstore_df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
print(top_sales)

top_sales.plot(kind='barh', color='skyblue')
plt.title("Top 10 Best-Selling Products")
plt.xlabel("Total Sales")
plt.gca().invert_yaxis()  
plt.show()
```

```python
bottom_loss = superstore_df.groupby('Product Name')['Profit'].sum().sort_values().head(10)
print(bottom_loss)

bottom_loss.plot(kind='barh', color='red')
plt.title("Bottom 10 Products by Loss")
plt.xlabel("Total Profit (Loss)")
plt.gca().invert_yaxis()
plt.show()
```

```python
best_seller_name = top_sales.index[0]
superstore_df[superstore_df['Product Name'] == best_seller_name][['Product Name','Sales','Profit']]
```
